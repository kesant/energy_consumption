```python
# Define the columns to keep
columns_to_keep = [
    'Date/Time',
    'Electricity:Facility [kW](Hourly)',
    'Fans:Electricity [kW](Hourly)',
    'Cooling:Electricity [kW](Hourly)',
    'Heating:Electricity [kW](Hourly)',
    'InteriorLights:Electricity [kW](Hourly)',
    'InteriorEquipment:Electricity [kW](Hourly)'
]
# Load and concatenate data from all files
dfs = []
for file in archivos:
    df = pd.read_csv(f"../data/raw/{file}", usecols=columns_to_keep)
    df["Date/Time"] = '2004 ' + df["Date/Time"]
    date_format = '%Y %m/%d %H:%M:%S'
    df["Date/Time"] = pd.to_datetime(df["Date/Time"], format=date_format, errors='coerce')
    match = re.match(r'^[^_]+', file)
    name = match.group(0)
    df["type_building"] = name
    dfs.append(df)
# # Combine all dataframes into one
combined_df = pd.concat(dfs, axis=0, ignore_index=True)
# # Set Date/Time as index
# combined_df.set_index('Date/Time', inplace=True)
```

```python
def calcular_series_promedio(dataframe):
    # Asegurarte de que la columna 'Date/Time' esté en formato datetime
    dataframe['Date/Time'] = pd.to_datetime(dataframe['Date/Time'])
    
    # Establecer la columna 'Date/Time' como índice
    dataframe.set_index('Date/Time', inplace=True)
    
    # Resamplear por hora y calcular el promedio
    dataframe_combinado = dataframe.iloc[:,:-1].resample('h').mean()
    return dataframe_combinado
```

```python
combined_df_cluster_0 = pd.concat(clustered_dfs[0], axis=0, ignore_index=True)
combined_df_cluster_0=calcular_series_promedio(combined_df_cluster_0)
combined_df_cluster_1 = pd.concat(clustered_dfs[1], axis=0, ignore_index=True)
combined_df_cluster_1=calcular_series_promedio(combined_df_cluster_1)
combined_df_cluster_2 = pd.concat(clustered_dfs[2], axis=0, ignore_index=True)
combined_df_cluster_2=calcular_series_promedio(combined_df_cluster_2)
combined_df_cluster_3 = pd.concat(clustered_dfs[3], axis=0, ignore_index=True)
combined_df_cluster_3=calcular_series_promedio(combined_df_cluster_3)
combined_df_cluster_4 = pd.concat(clustered_dfs[4], axis=0, ignore_index=True)
combined_df_cluster_4=calcular_series_promedio(combined_df_cluster_4)
```
